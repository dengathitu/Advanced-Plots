```python
# 1. Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Task 1: Import the dataset using Pandas
print("--- Task 1: Importing the dataset ---")
try:
    df = pd.read_csv('Iris Dataset.csv')
    # Let's inspect the column names to use them correctly
    print("Dataset imported successfully! Column names are:")
    print(df.columns)
except FileNotFoundError:
    print("Error: 'Iris Dataset.csv' not found.")
    exit()
print("-" * 40)
```

```python
# Task 2: Print the first five and last five rows
print("\n--- Task 2: First and Last Five Rows ---")
print("First 5 rows:")
print(df.head(5))
print("\nLast 5 rows:")
print(df.tail(5))
print("-" * 40)
```

```python
# Task 3: Check the detailed information of the dataset
print("\n--- Task 3: Detailed Dataset Information ---")
df.info()
print("-" * 40)
```

```python
# Task 4: Check if any null values are present
print("\n--- Task 4: Checking for Null Values ---")
print("Total null values in each column:")
print(df.isnull().sum())
print("-" * 40)
```

```python

# Task 5: Create a subset with data from row number 21 to 55
print("\n--- Task 5: Creating a Subset ---")
# .iloc selects rows by integer index. Row 21 is at index 20.
# The slice 20:55 includes index 20 up to, but not including, 55.
subset_df = df.iloc[20:55]
print(f"Subset from row 21 to 55 created. It has {subset_df.shape[0]} rows.")
print(subset_df.head())
print("-" * 40)
```

```python
# Task 6: Check details of the flower which has the highest Sepal Length
print("\n--- Task 6: Flower with Highest Sepal Length ---")
# Use idxmax() to get the index of the row with the maximum value.
# The actual column name is 'SepalLengthCm'.
highest_sepal_flower = df.loc[df['SepalLengthCm'].idxmax()]
print("Details of the flower with the highest sepal length:")
print(highest_sepal_flower)
print("-" * 40)
```

```python
# Task 7: Create a boxplot for all the features
print("\n--- Task 7: Generating Boxplot for All Features ---")
plt.figure(figsize=(10, 6))
# We drop non-numeric columns before plotting
sns.boxplot(data=df.drop(columns=['Id', 'Species']))
plt.title('Boxplot of Iris Dataset Features')
plt.ylabel('Centimeters')
plt.show()
print("-" * 40)
```

```python
# Task 8: Check the relationship between Sepal Length and Sepal Width
print("\n--- Task 8: Relationship between Sepal Length and Sepal Width ---")
plt.figure(figsize=(8, 6))
# A scatter plot is ideal for visualizing the relationship between two numeric variables.
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', data=df, hue='Species')
plt.title('Sepal Length vs. Sepal Width')
plt.grid(True)
plt.show()
print("-" * 40)
```

```python
# Task 9: Check the distribution of all the features
print("\n--- Task 9: Checking Feature Distributions ---")
# We will create a histogram for each numerical feature.
numerical_features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
df[numerical_features].hist(bins=20, figsize=(12, 8), grid=False)
plt.suptitle('Distribution of Iris Dataset Features')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
print("-" * 40)
```

```python
# Task 10: Create a count plot of the feature 'Species'
print("\n--- Task 10: Count Plot for Species ---")
plt.figure(figsize=(7, 5))
sns.countplot(x='Species', data=df)
plt.title('Number of Flowers per Species')
plt.ylabel('Count')
plt.show()
print("-" * 40)
```
